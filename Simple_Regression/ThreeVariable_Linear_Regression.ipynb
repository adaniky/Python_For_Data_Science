```python
#libs
import pandas  as pd
import numpy as np
```

```python
pr1 = pd.read_csv("X_train.csv", header=None)
pr1.columns = ["x1", "x2", "x3"] #we have three dependent variables. let's build a regression model
pr1.head()
```

```python
y1 = pd.read_csv("y_train.csv", header=None)
y1.columns = ["y"] #it is obvious that we have 4 arrays of data, or vectors, that are connected with each other. 
# If we assume that we are dealing with linear regression, the expression will be fair y =A + Bx1 + Cx2 + Dx3, 
# find А, В, С, D, constants for model building
y1.head()
```

Find the dependence using this system of equations
<img src="rege.JPG">

```python
matrix = pd.DataFrame( columns = ["x1", "x2", "x3", "x1x2", "x1x3", "x2x3", "x1^2", "x2^2", "x3^2", "y", "x1y","x2y", "x3y"] ) 
matrix["x1x2"] = pr1["x1"] * pr1["x2"]
matrix["x1x3"] = pr1["x1"] * pr1["x3"]
matrix["x2x3"] = pr1["x2"] * pr1["x3"]
matrix["x1"] = pr1["x1"]
matrix["x2"] = pr1["x2"]
matrix["x3"] = pr1["x3"]
matrix["x1^2"] = pr1["x1"] * pr1["x1"]
matrix["x2^2"] = pr1["x2"] * pr1["x2"]
matrix["x3^2"] = pr1["x3"] * pr1["x3"]
matrix["y"] = y1["y"]
matrix["x1y"] = pr1["x1"] * y1["y"]
matrix["x2y"] = pr1["x2"] * y1["y"]
matrix["x3y"] = pr1["x3"] * y1["y"]
sums = matrix.sum(axis=None)
```

```python
sums
```

```python
d = {'1':[200, sums['x1'], sums['x2'], sums['x3']], '2':[sums['x1'], sums['x1^2'], sums['x1x2'], sums['x1x3']],
     '3':[sums['x2'], sums['x1x2'], sums['x2^2'], sums['x2x3']], '4':[sums['x3'], sums['x1x3'], sums['x2x3'], sums['x3^2']]}
first_argument = pd.DataFrame(data=d) # Compute variable`s matrix
xsumm = np.linalg.inv(first_argument)# Compute the (multiplicative) inverse of a matrix with var`s sums.
```

```python
first_argument
```

```python
xsumm
```

```python
d2 = {'1':[sums['y'], sums['x1y'], sums['x2y'], sums['x3y']]} #Compute constant`s matrixs
second_argument = pd.DataFrame(data=d2)
xysumm = second_argument.as_matrix()
```

```python
second_argument
```

```python
Consts = np.dot(xsumm , xysumm)
Consts # found coefficients of depend between target and instrumental variables
```

```python
A = float(Consts[0])
B = float(Consts[1])
C = float(Consts[2])
D = float(Consts[3]) #name of constants
```

Found regression equation:<br>
y = 0,006+4,5*X1+1,3*X2-1,98*X3

```python
pr2 = pd.read_csv("X_test.csv", header=None) # download predictors for new target value
pr2.columns = ["x1", "x2", "x3"]
```
